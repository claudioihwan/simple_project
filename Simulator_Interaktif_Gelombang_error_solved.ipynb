```python
#@title Simulator Interaktif Gelombang

from matplotlib import pyplot as plt
import numpy as np
import math
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# initializing a figure in
# which the graph will be plotted
fig = plt.figure()

# marking the x-axis and y-axis
axis = plt.axes(xlim =(0, 10),
				ylim =(-10,10))

# initializing a line variable
line, = axis.plot([], [], lw = 3)
line2, = axis.plot([], [], lw = 2)

# data which the line will
# contain (x, y)
def init():
  line.set_data([],[])
  return line,
  line2.set_data([],[])
  return line2,


print('Gelombang I')
A=float(input('Masukkan amplitudo (m): '))
wi=float(input('Masukkan frekuensi sudut (derajat): '))
w=math.radians(wi)
#w*=np.pi
#fase=float(input('Masukkan fase: '))

print('Gelombang II')
A2=float(input('Masukkan amplitudo (m): '))
w2i=float(input('Masukkan frekuensi sudut (derajat): '))
w2=math.radians(w2i)
#w2*=np.pi
#fase2=float(input('Masukkan fase: '))

print(w, w2)

def animate(i):
  x=np.linspace(0,10,1000)

  y1=A*np.sin(w*(x+0.1*i))
  line.set_data(x,y1)


  y2=A2*np.sin(w2*(x+0.1*i))
  line2.set_data(x,y2)

  return line, line2


plt.xlabel('Arah rambat gelombang')
plt.ylabel('Arah simpangan gelombang')
plt.legend(["Gelombang I", "Gelombang II"], loc ="lower right")
anim = FuncAnimation(fig, animate, init_func = init,
					frames = 500, interval = 20, blit = True)


anim.save('continuousSineWave.gif', writer = 'pillow', fps = 50)
HTML(anim.to_html5_video())


```
